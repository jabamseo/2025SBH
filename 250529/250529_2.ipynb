```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation, rc
```

# 조화진동

```python
fig, ax = plt.subplots()

# 1. 그림 범위 설정
ax.set_xlim([-10, 10])

# 2. 점 초기화
T = 20
dt = 0.1
steps = int(T/dt)
m = 1.0
k = 1.0
x_init = 5
vx_init = 0
pos_init = ax.scatter(x_init, 0)

x = np.zeros(steps)
vx = np.zeros(steps)
for t in range(steps):
  if t == 0:
    x[t] = x_init
    vx[t] = vx_init
  else:
    a = -(k/m)*x[t-1]
    vx[t] = vx[t-1] + a*dt
    x[t] = x[t-1] + vx[t]*dt

# 3) 매 프레임마다 호출될 함수: 각 scatter를 새 위치(0, y[t])로 이동
def animate(t):
    pos_init.set_offsets((x[t], 0))
    return (pos_init,)

# 4) 애니메이션 객체 생성
ani = animation.FuncAnimation(
    fig,                  # 어느 Figure에 그릴지
    animate,              # 매 프레임마다 호출할 함수
    repeat=True,          # 끝나면 다시 처음부터 반복
    frames=steps,          # 총 프레임 수 = time 만큼
    interval=50           # 프레임 간 간격(밀리초)
)

# 5) 정적 그림 제거, jshtml로 표시
plt.close(fig)
rc('animation', html='jshtml')
ani
```

```python
tarray = np.linspace(0,T,int(T/dt))
plt.plot(tarray, x)
plt.show()
```

## x, y 성분 둘다 진동운동

```python
fig, ax = plt.subplots()

# 1. 그림 범위 설정
ax.set_xlim([-10, 10])
ax.set_ylim([-10, 10])

# 2. 점 초기화
T = 20
dt = 0.1
steps = int(T/dt)
m = 1.0
k = 1.0
x_init = 5
y_init = 5
vx_init = 0
vy_init = 0
pos_init = ax.scatter(x_init, y_init)

x = np.zeros(steps)
y = np.zeros(steps)
vx = np.zeros(steps)
vy = np.zeros(steps)
for t in range(steps):
  if t == 0:
    x[t] = x_init
    y[t] = x_init
    vy[t] = vx_init
    vx[t] = vx_init
  else:
    ax = -(k/m)*x[t-1]
    ay = -(k/m)*y[t-1]
    vx[t] = vx[t-1] + ax*dt
    vy[t] = vy[t-1] + ay*dt
    x[t] = x[t-1] + vx[t]*dt
    y[t] = y[t-1] + vy[t]*dt

# 3) 매 프레임마다 호출될 함수: 각 scatter를 새 위치(0, y[t])로 이동
def animate(t):
    pos_init.set_offsets((x[t], y[t]))
    return (pos_init,)

# 4) 애니메이션 객체 생성
ani = animation.FuncAnimation(
    fig,                  # 어느 Figure에 그릴지
    animate,              # 매 프레임마다 호출할 함수
    repeat=True,          # 끝나면 다시 처음부터 반복
    frames=steps,          # 총 프레임 수 = time 만큼
    interval=50           # 프레임 간 간격(밀리초)
)

# 5) 정적 그림 제거, jshtml로 표시
plt.close(fig)
rc('animation', html='jshtml')
ani
```

# 감쇠진동

```python
fig, ax = plt.subplots()

# 1. 그림 범위 설정
ax.set_xlim([-10, 10])

# 2. 점 초기화
T = 20
dt = 0.1
steps = int(T/dt)
m = 1.0
k = 1.0
c = 0.2
x_init = 5
vx_init = 0
pos_init = ax.scatter(x_init, 0)

x = np.zeros(steps)
vx = np.zeros(steps)
for t in range(steps):
  if t == 0:
    x[t] = x_init
    vx[t] = vx_init
  else:
    a = -(k/m)*x[t-1] - (c/m)*vx[t-1]
    vx[t] = vx[t-1] + a*dt
    x[t] = x[t-1] + vx[t]*dt

# 3) 매 프레임마다 호출될 함수: 각 scatter를 새 위치(0, y[t])로 이동
def animate(t):
    pos_init.set_offsets((x[t], 0))
    return (pos_init,)

# 4) 애니메이션 객체 생성
ani = animation.FuncAnimation(
    fig,                  # 어느 Figure에 그릴지
    animate,              # 매 프레임마다 호출할 함수
    repeat=True,          # 끝나면 다시 처음부터 반복
    frames=steps,          # 총 프레임 수 = time 만큼
    interval=50           # 프레임 간 간격(밀리초)
)

# 5) 정적 그림 제거, jshtml로 표시
plt.close(fig)
rc('animation', html='jshtml')
ani
```

```python
tarray = np.linspace(0,T,int(T/dt))
plt.plot(tarray, x)
plt.show()
```

# 등속 원운동

```python
T = 20
dt = 0.1
steps = int(T/dt)
A = 5
w = 1

# Prepare arrays x, y, z
tarray = np.linspace(0,T,steps)
x = A * np.cos(w * tarray)
y = A * np.sin(w * tarray)

ax = plt.figure().add_subplot(projection='3d')
ax.plot(tarray, x, y, label='parametric curve')
ax.legend()

plt.show()
```

```python
import numpy as np
import plotly.graph_objects as go

#--- 파라미터 및 데이터 생성 ---
T = 20
dt = 0.01
steps = int(T/dt)
A = 5
w = 10

# Prepare arrays x, y, z
tarray = np.linspace(0,T,steps)
x = A * np.cos(w * tarray)
y = A * np.sin(w * tarray)
z = tarray  # 시간 축을 z 로 씁니다.

#--- Plotly 3D 궤적 그리기 ---
fig = go.Figure(
    data=go.Scatter3d(
        x=z,         # 시간
        y=x,         # x(t)
        z=y,         # y(t)
        mode='lines',
        line=dict(width=4, color='blue'),
        name='parametric curve'
    )
)

# 축 레이블과 레이아웃
fig.update_layout(
    scene=dict(
        xaxis_title='t',
        yaxis_title='x(t)',
        zaxis_title='y(t)',
        aspectmode='auto'
    ),
    width=700, height=600,
    title='인터랙티브 3D Parametric Curve'
)

fig.show()
```
